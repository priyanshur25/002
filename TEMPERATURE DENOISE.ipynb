```python
#visualizing the dataset
import pandas as pd
df=pd.read_csv('Complete_Details(1-552) (1).csv')

print(df.shape)

print(df.head())
```

```python
import matplotlib.pyplot as plt
plt.plot(df['Temperature'])
plt.title('temperature')
plt.show()
```

```python

import numpy as np
from scipy.signal import medfilt



# our signal data is in a column named 'Temperature'
signal_data = df['Temperature']

# Apply a median filter to remove spikes
window_size = 3  # we can adjust this parameter based on the characteristics of your spikes
filtered_signal = medfilt(signal_data, kernel_size=window_size)

# Plot the original and filtered signals
plt.figure(figsize=(10, 6))

plt.subplot(2, 1, 1)
plt.plot(signal_data)
plt.title('Original Signal')
plt.xlabel('Time')
plt.ylabel('Signal Amplitude')

plt.subplot(2, 1, 2)
plt.plot(filtered_signal)
plt.title('Filtered Signal')
plt.xlabel('Time')
plt.ylabel('Signal Amplitude')

plt.tight_layout()
plt.show()

#snr
signal_power = np.sum(filtered_signal**2) / len(filtered_signal)
noise_power = np.sum((signal_data - filtered_signal)**2) / len(signal_data)

# Calculate SNR in decibels (dB)
snr = 10 * np.log10(signal_power / noise_power)
print(f"Signal-to-Noise Ratio (SNR): {snr:.2f} dB")
```
